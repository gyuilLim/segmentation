```python
import json
import matplotlib.pyplot as plt
import os
```

```python
train_log = '../save/unetpp_train_log.json'
test_log = '../save/unetpp_test_log.json'


with open(train_log, 'r') as file :
    train_log = json.load(file)

with open(test_log, 'r') as file :
    test_log = json.load(file)

train_loss_list = train_log['train_losses']
train_dice_list = train_log['train_dice_scores']
train_iou_list = train_log['train_iou_scores']

plt.figure(figsize=(12, 5))

plt.plot(train_loss_list, label = 'train loss')
plt.plot(train_dice_list, label = 'train_dice')
plt.plot(train_iou_list, label = 'train_iou')
```
