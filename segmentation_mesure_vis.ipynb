```python
import json
import matplotlib.pyplot as plt
import os
```

```python
test_iou = '../save/unetpp_testiou.json'
test_dice = '../save/unetpp_testdice.json'

with open(test_iou, 'r') as file :
    test_iou_info = json.load(file)

with open(test_dice, 'r') as file :
    test_dice_info = json.load(file)


test_iou_list = [loss for loss in test_iou_info]
test_dice_list = [loss for loss in test_dice_info]

test_x = [i * 10 for i in range(len(test_dice_list))]

plt.figure(figsize=(12, 5))

plt.plot(test_x, test_iou_list, label='test iou')
plt.plot(test_x, test_dice_list, label='test dice')
plt.legend()
```

```python
import numpy as np

iou_idx = np.argmax(test_iou_list)
dice_idx = np.argmax(test_dice_list)
print(f"{test_iou_list[iou_idx]:.4f}({iou_idx*10}), {test_dice_list[dice_idx]:.4f}({dice_idx*10})")
```
